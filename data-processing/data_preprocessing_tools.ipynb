## Importing the dataset

```
dataset = pd.read_csv('Data.csv') #Read data from the CSV with pandas lib
x = dataset.iloc[:,0:-1].values  # materix of the feature. iloc => help to locate index iloc[rows,columns] => rows "start:end",
y = dataset.iloc[:,-1].values  # Depending variables, alternative=> dataset.iloc[:,-1:]
```

## Taking care of missing data

```
# Missing data will make model create problems in model learning.
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(missing_values=np.nan,strategy='mean') #Replace missing values of nan with mean values other values.
imputer.fit(x[:,1:3]) # Fit will required all the columns with numeric values
x[:,1:3]= imputer.transform(x[:,1:3]) #Assing the values to the same columns

```
